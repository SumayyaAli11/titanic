```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
```

```python
sns.barplot(x='Pclass', y='Age', data=df,hue="Sex")
```

```python
sns.barplot(x='Pclass', y='Fare', data=df)
```

```python
sns.boxplot(x='Sex', y='Age', data=df)
#insight:outliers in males
```

```python
sns.boxplot(x='Sex', y='Age', data=df,hue='Survived')
#insight:In males,younger population has more survival and in females, older population has more survival
```

```python
sns.distplot(df[df['Survived']==0]['Age'],hist=False)
sns.distplot(df[df['Survived']==1]['Age'],hist=False)
```

insight:Blue curve:those who died,orange:those who survived.If you are a child,probability of survival > probability of dying

```python
pd.crosstab(df['Pclass'],df['Survived'])
```

insight:more survival in P1 class

```python
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
#lighter shade:higher value,in P3 class, only less people is survived
```

```python
df.groupby('Pclass')['Survived'].mean()*100
```

```python
(df.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
df.groupby('Embarked')['Survived'].mean()*100
```

```python
pd.crosstab(df['SibSp'],df['Survived'])
```

```python
sns.clustermap(pd.crosstab(df['SibSp'],df['Survived']))
#insight:SibSp 3 and 4 is similar
```
